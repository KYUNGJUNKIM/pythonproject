```python
#ie_statistics hw3 #8-1

import math
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import norm

xbar = 65
mu = 60
sigma = 9.5
alpha = 0.05
n = 15
se = sigma/math.sqrt(n)
z0 = (xbar-mu)/(sigma/math.sqrt(n))

#
standard_norm = norm(0,1)
z = standard_norm.ppf(0.95)

if z0 >= z:
    print("H0 rejected")
else:
    print("H0 accepted")

#2
def r_beta(x, n, alpha, sigma):
    se = sigma/math.sqrt(n)
    rv = norm(loc = mu, scale = se)
    q = rv.isf(alpha)
    rc = norm(loc = x, scale = se)
    beta = rc.cdf(q)
    return beta

k = np.arange(60, 70, 0.1)
n = [1, 2, 5, 10, 15, 20, 30, 40, 50, 100]
for i in n:
     plt.plot(k, r_beta(k, i, alpha, sigma), label="OC curve")
plt.xlabel('True Mean')
plt.ylabel('Probaility of type 2 error: beta')
plt.show()

#
t = norm.ppf(0.975)
if z0 >= t:
    print("H0 rejected")
else:
    print("H0 accepted")

#2
def r_beta(x, n, alpha, sigma):
    se = sigma/math.sqrt(n)
    rv = norm(loc = mu, scale = se)
    acc = [alpha/2, 1-alpha/2]
    q = rv.ppf(acc)
    rc = norm(loc = x, scale = se)
    beta = rc.cdf(q[1])-rc.cdf(q[0])
    return beta

k = np.arange(50, 70, 0.1)
n = [1, 2, 5, 10, 15, 20, 30, 40, 50, 100]
for i in n:
     plt.plot(k, r_beta(k, i, alpha, sigma), label="OC curve")
plt.xlabel('True Mean')
plt.ylabel('Probaility of type 2 error: beta')
plt.show()
```
